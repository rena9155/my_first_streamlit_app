```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
hund_df_raw = pd.read_csv("/data/processed/20200306_hundehalter.csv")
hund_df_raw.head()
```

```python
filtered_df = hund_df_raw[hund_df_raw['RASSE1_MISCHLING'].notnull()]
hund_df_raw['RASSENTYP'].unique()
```

```python
dog_df_raw = hund_df_raw.rename(columns = {
    "HALTER_ID": 'Holder_ID',
    "ALTER": 'Age',
    "GESCHLECHT":'Holder_Sex',
    "STADTKREIS":'District',
    "STADTQUARTIER":'City_Code',
    "RASSE1":'Race1',
    "RASSE1_MISCHLING":'Race1_Hybrid',
    "RASSE2":'Race2',
    "RASSE2_MISCHLING":"Race2_Hybrid",
    "RASSENTYP":'Race',
    "GEBURTSJAHR_HUND":'Dog_Birth_Year',
    "GESCHLECHT_HUND":'Dog_sex',
    "HUNDEFARBE":'Dog_Color'
})
dog_df_raw
```

```python
dog_df_raw['Dog_Age'] = 2023 - pd.to_numeric(dog_df_raw['Dog_Birth_Year'])
dog_df_raw['Age'].unique()
```

```python
age_group_dict = {"11-20":1,"21-30":2,"31-40":3,"41-50":4,"51-60":5,"61-70":6,"71-80":7,"81-90":8,"91-100":9}
dog_df_raw['Owner_Age_Group']=dog_df_raw['Age'].map(age_group_dict)
dog_df_raw['Owner_Age_Group']=dog_df_raw['Owner_Age_Group'].astype(int)
dog_df_raw
```

```python
dog_df_raw.to_csv('../data/processed/Dog_data.csv')
```
